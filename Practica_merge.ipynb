Parte 2 — Merge

3. Crear orders_customers

Realiza un merge entre:

orders

customers

Usando la columna:

customer_id

Guarda el resultado en:

orders_customers

```python
orders_customers = pd.merge(
    orders,
    customers,
    on = "customer_id",
    how = "inner"
)
orders_customers
```

5. Crear items_products

Realiza un merge entre:

items

products

usando la columna:

product_id

Guarda el resultado en:

items_products

```python
items_products = pd.merge(
    items,
    products,
    on = "product_id",
    how = "inner"
)
items_products
```
